```python
now = dt.now()
current_time = now.strftime("%d-%m-%Y %H-%M-%S")
current_time
```

```python
# scraping through multiple pages to get the listing urls and then appending them to a dataframe
masterurl = 'https://www.magicbricks.com/property-for-sale/residential-real-estate?&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment,Service-Apartment,Residential-House,Villa,Residential-Plot&cityName='+landmark+'&mbTrackSrc=tabChange&page=1&category=S'
masterresponse = requests.get(masterurl)
mastersoup = BeautifulSoup(masterresponse.content, "html.parser")


maxno = 0

for elements in mastersoup.find_all(['div'],class_ = "SRmTab"):
    for subelements in elements.find_all(['a'],href = "javascript:void(0);"):
        if 'Properties' in subelements.text:
            for mainelements in subelements.find_all(['span'],itemprop = "name"):
                maxno = int(mainelements.text[1:len(mainelements.text)-1])



urlframe3 = []

pagelist = range(1,15000)
for page in pagelist:
    
    if maxno <= 0:
        break
    
    url3 = 'https://www.magicbricks.com/property-for-sale/residential-real-estate?&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment,Service-Apartment,Residential-House,Villa,Residential-Plot&cityName='+landmark+'&mbTrackSrc=tabChange&page='+str(page)+'&category=S'
    response3 = requests.get(url3)
    soup3 = BeautifulSoup(response3.content, "html.parser")
    
    for url in soup3.find_all(['div','meta'],class_ = "m-srp-card SRCard"):
        a = []
        urlframe3.append(url["data-code"][23:len(url["data-code"])-18]+"_"+str(page))
        a.append(url["data-code"][23:len(url["data-code"])-18]+"_"+str(page))
        maxno = maxno - len(a)
        print(maxno, page,url["data-code"][23:len(url["data-code"])-18])
```

```python
listofurlsv2 = pd.DataFrame(urlframe3)
listofurlsv2.to_csv("urlframe_"+str(current_time)+".csv", index=False, header=False)
```
