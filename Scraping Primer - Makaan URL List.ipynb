```python
now = dt.now()
current_time = now.strftime("%d-%m-%Y %H-%M-%S")
current_time
```

```python
# scraping the list of URLs and appending them to a dataframe


masterurl = 'https://www.makaan.com/'+landmark+'-residential-property/buy-property-in-'+landmark+'-city?budget=,&page=1'
masterresponse = requests.get(masterurl)
mastersoup = BeautifulSoup(masterresponse.content, "html.parser")

urllist = []
maxno = 0
for elements in mastersoup.find_all(['div'],class_='f-count-wrap'):
    for subelements in elements.find_all(['strong'], class_ = "val-count"):
        maxno = int(subelements.text.replace(',',''))
        
print(maxno)
    

pagelist = range(1,3501)

for pageno in pagelist:
    
    if maxno <= 0:
        break

    masterurl = 'https://www.makaan.com/'+landmark+'-residential-property/buy-property-in-'+landmark+'-city?budget=,&page='+str(pageno)
    masterresponse = requests.get(masterurl)
    mastersoup = BeautifulSoup(masterresponse.content, "html.parser")



    for elements in mastersoup.find_all(['a','target'], itemprop="url", class_="typelink"):
    #for elements2 in elements.find_all('a'):
        a=[]
        a.append(elements["href"])
        urllist.append(elements["href"])
        maxno = maxno-1
        print(maxno, pageno)

urllist
```

```python
# Checking for duplicates
urllist_pd = pd.DataFrame(urllist)
print(len(urllist_pd),len(urllist_pd.drop_duplicates()))
```

```python
listofurls = pd.DataFrame(urllist_pd)
listofurls.to_csv("urlframe_"+str(current_time)+".csv", index=False, header=False)
```
